```python
def calc_cagr2(co):
    try:
        df = pd.read_csv(f'../Cleaned/NSE/BS/{co}_BS.csv', index_col=0)

        cagr = {}
        for i,d in enumerate(df.columns):
            if df[d].name.isupper() and len(df[d].name) > 8:
                pass
            else:
                if float(df.at['Mar 21',d]) != 0 and float(df.at['Mar 17',d]) != 0:
                    n = (float(df.at['Mar 21',d]) / float(df.at['Mar 17',d]) )**(1/5) - 1
                    cagr[f'cagr_{d[:8]}'] = n
                elif float(df.at['Mar 20',d]) != 0 and float(df.at['Mar 17',d]) != 0:
                    cagr[f'cagr_{d[:8]}'] = (float(df.at['Mar 21',d]) / float(df.at['Mar 17',d]) )**(1/4) - 1
                elif float(df.at['Dec 20',d]) != 0 and float(df.at['Dec 16',d]) != 0:
                    cagr[f'cagr_{d[:8]}'] = (float(df.at['Dec 20',d]) / float(df.at['Dec 16',d]) )**(1/4) - 1
                
                else:
                    pass

        cagrdf = pd.DataFrame(data=[cagr.values()], columns=cagr.keys())
        cagrdf.to_csv(f'../Cleaned/NSE/CAGR/{co}_cagr.csv')
        print(f'data printed for {co}')
    except FileNotFoundError:
        print(f'file not found for {co}')   
```

```python
def calc_cagr(co):
    try:
        df = pd.read_csv(f'../Cleaned/NSE/BS/{co}_BS.csv', index_col=0)

        cagr = {}
        for d in df.columns:
            if df[d].name.isupper() and len(df[d].name) > 8:
                pass
            else:
                series = df[d]
                fi = series[0]
                li = series[-1]
                cagr[f'cagr_{d[:8]}'] = (fi/li)**(1/len(series)) -1
        cagrdf = pd.DataFrame(data=[cagr.values()], columns=cagr.keys())
        cagrdf.to_csv(f'../Cleaned/NSE/CAGR/{co}_cagr.csv')
        print(f'data printed for {co}')
    except FileNotFoundError:
        print(f'file not found for {co}') 
                        
```
