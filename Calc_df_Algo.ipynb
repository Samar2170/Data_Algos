```python
import pandas as pd
import os
```

```python
def calc_df(co):
    try:
        dfbs = pd.read_csv(f"../NSE/BS/NonFinancial/{co}_stnd_BS.csv")
        dfpl = pd.read_csv(f"../NSE/PL/NonFinancial/{co}_stnd_PL.csv")
        if len(dfbs.columns) == 8 and len(dfpl.columns) == 8:
            dfbs.drop(['Unnamed: 0', '6'],axis=1, inplace=True)
            dfpl.drop(['Unnamed: 0', '6'],axis=1, inplace=True)

            dfbs.set_index('0', inplace=True)
            dfpl.set_index('0', inplace=True)

            dfbs.columns = dfbs.iloc[0]
            dfpl.columns = dfpl.iloc[0]

            dfbs = dfbs.transpose()
            dfpl = dfpl.transpose()

            dfbs.columns = dfbs.columns.fillna('to_drop')
            dfbs.drop('to_drop', axis=1, inplace=True)

            dfpl.columns = dfpl.columns.fillna('to_drop')
            dfpl.drop('to_drop', axis=1, inplace=True)


            cd = dfbs.filter(regex=("Balance Sheet"),axis=1)
            dfbs.drop(cd, axis=1, inplace=True)

            cd2 = dfpl.filter(regex=("Profit & Loss"), axis=1)
            dfpl.drop(cd2,axis=1, inplace=True)

            dfbs.replace(to_replace='--', value=0, inplace=True)
            dfpl.replace(to_replace='--', value=0, inplace=True)

            dfbs.fillna(0, inplace=True)
            dfpl.fillna(0, inplace=True)

            indx = dfpl.index.name
            df = dfbs.join(dfpl, rsuffix="_pl")
            cols = []
            for d in df.columns:
                cols.append(d)
                
            for d in df.columns:
                df[d] = df[d].astype(float)


            df['EBITDA'] = (df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] + df['Finance Costs'] + df['Depreciation And Amortisation Expenses'])
            df['EBIT'] = df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] + df['Finance Costs']
            df['Total Debt'] = df['Long Term Borrowings'] +  df['Short Term Borrowings']
            
            df['DE'] = (df['Long Term Borrowings'] +  df['Short Term Borrowings']  )  /df['Total Reserves and Surplus']
            df['R/P'] = df['Trade Receivables'] / df['Trade Payables']
            df['CR'] = df['Total Current Assets'] / df['Total Current Liabilities']
            df['CCR'] = df['Cash And Cash Equivalents'] / df['Total Current Assets']
            df['Eq_Assets'] = df['Total Shareholders Funds'] / df['Total Capital And Liabilities']
            df['FA_Assets'] = df['Fixed Assets'] / df['Total Assets']
            df['NC_Assets'] = df['Total Non-Current Assets'] / df['Total Assets'] 
            df['C_Assets'] = df['Total Current Assets'] / df['Total Assets'] 

            df['EBITDA_Mgn'] = (df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] + df['Finance Costs'] + df['Depreciation And Amortisation Expenses']) / df['Total Revenue']
            df['EBIT_Mgn'] = (df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] + df['Finance Costs'] ) / df['Total Revenue']
            df['PBT_Mgn'] = df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] / df['Total Revenue']
            df['NPM'] = df['Profit/Loss For The Period']/df['Total Revenue']
            df['OPR_Rev'] = df['Total Operating Revenues'] / df['Total Revenue']
            df['RONW'] = df['Profit/Loss For The Period'] / df['Total Shareholders Funds']
            df['ROA'] = df['Profit/Loss For The Period'] / df['Total Assets']
            df['ITR'] = df['Total Revenue']/ df['Inventories']
            df['cred_sales_ratio'] =  df['Total Revenue']/ df['Trade Receivables']
            
            dfm = df.copy()
            
            for c  in cols:
                dfm.drop(c, axis=1, inplace=True)

            df.to_csv(f'../Cleaned/NSE/PL_BS/{co}_plbs.csv')
            dfbs.to_csv(f'../Cleaned/NSE/BS/{co}_BS.csv')
            dfpl.to_csv(f'../Cleaned/NSE/PL/{co}_PL.csv')
            dfm.to_csv(f'../Cleaned/NSE/Ratios/{co}_ratios.csv')
            print(f'data saved for {co}')
        else:
            print(f'inconsistent data for {co}')
    except FileNotFoundError:
        print(f'file not found for {co}')
```

```python
co_df = pd.read_csv('nifty500list.csv')
for i,r in co_df.iterrows():
    calc_df(r['Symbol'])
```
