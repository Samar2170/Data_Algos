```python
ffs = 0
svn = 0
eght = 0
incs = 0
for i,r in df.iterrows():
    try:
        co = r['Symbol']
        dff = pd.read_csv(f'NSE/BS/{co}_stnd_BS.csv')
        if len(dff.columns) == 8:
            eght+=1
        elif len(dff.columns) == 7:
            svn+=1
        else:
            incs+=1
        print(len(dff.columns))
    except FileNotFoundError:
        ffs +=1
        print('File not found')
print('files not f -- ' + str(ffs)) 
print(f'8 - {str(eght)}')
print(f'7 - {str(svn)}')
print(f'incs - {str(incs)}')
```
