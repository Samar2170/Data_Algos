```python
import pandas as pd
```

```python
df = pd.read_csv('Machine_Learning/prices_stocksASIANPAINT10yr.csv')
```

```python
df['max_dv'] = (df['high']/df['low']) - 1
df['close_dv'] = (df['close']/df['open']) - 1
```

```python
df.head()
```

```python
for i in range(1, len(df)):
    df.loc[i, 'mov_1d'] = (df.loc[i, 'close'] / df.loc[i-1, 'close']) - 1
    df.loc[i, 'vol_1d'] = (df.loc[i, 'volume'] / df.loc[i-1, 'volume']) - 1
```

```python
df.head()
```

```python
df['deta_mov_vol'] = df['mov_1d'] / df['vol_1d'] - 1
```
