```python
import os
import pandas as pd
os.getcwd()
```

```python
dfb = pd.read_csv('NSE/BS/NonFinancial/TCS_stnd_BS.csv')
dfb.head()
```

```python
dfp = pd.read_csv('NSE/PL/NonFinancial/TCS_stnd_PL.csv')
dfp.head()
```

```python
dfb.drop(['Unnamed: 0','6'],axis=1, inplace=True)
dfp.drop(['Unnamed: 0','6'],axis=1, inplace=True)
```

```python
dfb = dfb.set_index('0')
dfp = dfp.set_index('0')
```

```python
dfb.head()
```

```python
dfp.head()
```

```python
dfb.columns = dfb.iloc[0]
dfp.columns = dfp.iloc[0]
```

```python
dfp.head()
```

```python
dfb = dfb.transpose()
dfp = dfp.transpose()
```

```python
dfb.columns = dfb.columns.fillna('to_drop')
dfb.drop('to_drop', axis = 1, inplace = True)
dfb.head()
```

```python
dfp.columns = dfp.columns.fillna('to_drop')
dfp.drop('to_drop', axis = 1, inplace = True)
dfp.head()
```

```python
cd = dfb.filter(regex=("Balance Sheet*"), axis=1)
dfb.drop(cd,axis=1,inplace=True)
dfb.head()
```

```python
cd = dfp.filter(regex=("Profit & Loss*"), axis=1)
dfp.drop(cd,axis=1,inplace=True)
dfp.head()
```

```python
dfb.replace(to_replace='--', value=0, inplace=True)
dfp.replace(to_replace='--', value=0, inplace=True)
```

```python
dfb.fillna(0, inplace=True)
dfp.fillna(0, inplace=True)
```

```python
dfp.columns
```

```python
dfb.columns
```

```python
indx = dfp.index.name
df = dfb.join(dfp, rsuffix='_pl')
df
```

```python
for d in df.columns:
    df[d] = df[d].astype(float)
```

```python
df['DE'] = (df['Long Term Borrowings'] +  df['Short Term Borrowings']  )  /df['Total Reserves and Surplus']
df['R/P'] = df['Trade Receivables'] / df['Trade Payables']
df['CR'] = df['Total Current Assets'] / df['Total Current Liabilities']
df['CCR'] = df['Cash And Cash Equivalents'] / df['Total Current Assets']
df['Eq_Assets'] = df['Total Shareholders Funds'] / df['Total Capital And Liabilities']
df['FA_Assets'] = df['Fixed Assets'] / df['Total Assets']
df['NC_Assets'] = df['Total Non-Current Assets'] / df['Total Assets'] 
df['C_Assets'] = df['Total Current Assets'] / df['Total Assets'] 
```

```python
df.columns
```

```python
for d in df.columns:
    df[d] = df[d].astype(float)
df['EBITDA_Mgn'] = (df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] + df['Finance Costs'] + df['Depreciation And Amortisation Expenses']) / df['Total Revenue']
df['EBIT_Mgn'] = (df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] + df['Finance Costs'] ) / df['Total Revenue']
```

```python
df['PBT_Mgn'] = df['Profit/Loss Before Exceptional, ExtraOrdinary Items And Tax'] / df['Total Revenue']
```

```python
df['NPM'] = df['Profit/Loss For The Period']/df['Total Revenue']
```

```python
df['OPR_Rev'] = df['Total Operating Revenues'] / df['Total Revenue']
```

```python
df.head()
```

```python
df['RONW'] = df['Profit/Loss For The Period'] / df['Total Shareholders Funds']
df['ROA'] = df['Profit/Loss For The Period'] / df['Total Assets']
```

```python
df['ITR'] = df['Total Revenue']/ df['Inventories']
```

```python
df['cred_sales_ratio'] =  df['Total Revenue']/ df['Trade Receivables']
```
