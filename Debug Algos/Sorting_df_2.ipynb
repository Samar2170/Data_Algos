```python
import pandas as pd
import os
import shutil
```

```python
df = pd.read_csv('nifty500list.csv')
df.head()
```

```python
nfa = []
for i,r in df.iterrows():
    try:
        if r['Industry'] == 'FINANCIAL SERVICES':
            co = r['Symbol']
            target = (r'NSE/PL/Financial/'+co+'_stnd_PL.csv')
            d = pd.read_csv(target)
        else:
            co = r['Symbol']
            target = (r'NSE/PL/NonFinancial/'+co+'_stnd_PL.csv')
            d = pd.read_csv(target)
    except FileNotFoundError:
        co = r['Symbol']
        dic = {'co':co, 'industry':r['Industry'], 'name':r['Company Name']}
        nfa.append(dic)
```
