```python
df.drop('IAssets_gth',1, inplace=True)
for i in range(len(df)-1):
    if i == 4:
        df.loc[i,'IAssets_gth'] = 0.0
    else:    
        df.loc[i,'IAssets_gth'] = df.loc[i,'Intangible Assets'] / df.loc[i+1,'Intangible Assets']
    

df['IAssets_gth']
```
