```python
import pandas as pd
import os
```

```python
import numpy as np
def price_val_df():
    df = pd.read_csv(f'nifty500_enhanced.csv', index_col=0)
    df = df.copy()
    df['price'] = np.nan
    for i,r in df.iterrows():
        co = r['Symbol']
        pr_df = pd.read_csv(f'../corp_prices/{co}_10yr_price_md.csv', index_col=0)
        pr_df['turnover'] = pr_df['close']*pr_df['volume']
        dv_avg = pr_df['close_dv'].mean()
        trn_avg = pr_df['turnover'].mean()
        
        pr_df = pr_df.iloc[-1]
        df.at[i,'price'] = pr_df['close']
        df.at[i,'dv_avg'] = dv_avg
        df.at[i,'trn_avg'] = trn_avg
        print(f'row added for {co}')
    df.to_csv('nifty500_enp.csv')
        
        
        
```

```python
price_val_df()
```
