```python
import pandas as pd
import os
```

```python
def calc_BS_comps(co):
    try:
        df_ip = pd.read_csv(f'../Cleaned/NSE/BS/{co}_BS.csv')
        drop_cols = []
        for d in df_ip.columns:
            drop_cols.append(d)
        df = df_ip.copy()
        
        asset_cols = ['Fixed Assets', 'Total Non-Current Assets','Cash And Cash Equivalents',
               'Total Current Assets', 'Non-Current Investments', 'Current Investments','Inventories']
        
        liab_cols = ['Trade Payables','Total Shareholders Funds','Long Term Borrowings','Short Term Borrowings']
        
        for a in asset_cols:
            df[f'assets_{a[:8]}'] = df[a] / df['Total Assets']
        
        for l in liab_cols:
            df[f'liab_{l[:8]}'] = df[l] / df['Total Capital And Liabilities']
        
        for d in drop_cols:
            df.drop(d, axis=1, inplace=True)
        
        df.to_csv(f'../Cleaned/NSE/BS_Comp/{co}_BS_comp.csv')
        print(f'data saved for {co}')
    except FileNotFoundError:
        print(f'file not found for {co}')
        
```

```python
co_df = pd.read_csv('nifty500list.csv')
for i,r in co_df.iterrows():
    calc_BS_comps(r['Symbol'])
```
