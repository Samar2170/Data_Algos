```python
import json
direc = {}
def get_valuation(co):
    try:
        df = pd.read_csv(f'../KeyFins/{co}_KF.csv', index_col=0)
        df = df.transpose()
        try:
            eps = df.at[f'{co}.NS','trailingEps']
            bv = df.at[f'{co}.NS','bookValue']
            insider_h = df.at[f'{co}.NS','heldPercentInsiders']
            inst_h = df.at[f'{co}.NS','heldPercentInstitutions']
            df['float_pc'] = float(df.at[f'{co}.NS','floatShares'])/ float(df.at[f'{co}.NS','sharesOutstanding'])
            float_pc = df.at[f'{co}.NS','float_pc']
            dic = {'eps':float(eps),'bv':float(bv), 'insider_h':float(insider_h)*100, 'inst_h':float(inst_h)*100, 'float_pc':float_pc }
            dic = json.dumps(dic)
            dic = json.loads(dic)

            with open(f'../Supplm/{co}.json', 'w') as f:
                json.dump(dic, f)
            print(f'data for {co} saved')
        except Error as e: 
            print('data not formal  '+ str(e))
    except FileNotFoundError:
        print('file not found')    
```

```python
import numpy as np
df['eps'] = np.nan
df['bv']=np.nan
df['insider_h']=np.nan
df['inst_h']=np.nan
df['float_pc']=np.nan

def write_corp_df():
    for i,r in df.iterrows():
        co = r['Symbol']
        try:
            j = open(f'../Supplm/{co}.json')
            dic = json.load(j)
            df.at[i,'eps'] = dic['eps']
            df.at[i,'bv'] = dic['bv']
            df.at[i,'insider_h'] = dic['insider_h']
            df.at[i,'inst_h'] = dic['inst_h']
            df.at[i,'float_pc'] = dic['float_pc']
            print(f'row added for {co}')
            
        except FileNotFoundError:
            print('file not found')
    df.to_csv('nifty500_enhanced.csv')       
```
