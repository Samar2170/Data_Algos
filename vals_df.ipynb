```python
def get_valuation(co):
    try:
        df = pd.read_csv(f'../KeyFins/{co}_KF.csv', index_col=0)
        df = df.transpose()
        try:
            eps = df.at[f'{co}.NS','trailingEps']
        except:
            eps = 0
        try:
            bv = df.at[f'{co}.NS','bookValue']
            try:
                insider_h = df.at[f'{co}.NS','heldPercentInsiders']
            except:
                insider_h = 0
            try:
                inst_h = df.at[f'{co}.NS','heldPercentInstitutions']
            except:
                inst_h = 0

            try:
                df['float_pc'] = float(df.at[f'{co}.NS','floatShares'])/ float(df.at[f'{co}.NS','sharesOutstanding'])
                float_pc = df.at[f'{co}.NS','float_pc']
            except:
                float_pc = 0
            try:
                dic = {'eps':float(eps),'bv':float(bv), 'insider_h':float(insider_h)*100, 'inst_h':float(inst_h)*100, 'float_pc':float_pc }
                dic = json.dumps(dic)
                dic = json.loads(dic)
                with open(f'../Supplm2/{co}.json', 'w') as f:
                    json.dumps(dic, f)
                print(f'data for {co} saved')    
                return dic
            except:
                pass
        except:
            print('data not uniform')
    except FileNotFoundError:
        print('file not found')    



```

```python
for a in ar:
    get_valuation(a)
```
